```python
top_scorers=ipl_data.groupby('batter')['total_runs'].sum().sort_values(ascending=False).head(10)
```

```python
top_wicket_takers=ipl_data.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(ipl_data['total_runs'],bins=20,kde=True)
plt.title('Distribution of Runs scored')
plt.xlabel('Runs Scored')
plt.ylabel('frequency')
plt.show()
```

```python
top_scorers.plot(kind='bar')
plt.title('Top 10 Runs Scorer')
plt.xlabel('Player')
plt.ylabel('Total Runs')
plt.show()
```

```python
top_wicket_takers.plot(kind='bar',color='red')
plt.title('Top 10 Wicket takers')
plt.xlabel('Player')
plt.ylabel('Total Wickets')
plt.show()
```

```python
matches_per_year=ipl_data.groupby('match_id').nunique()
```

```python
matches_per_year
```

```python
matches_per_year.plot(kind='line',marker='1')
plt.title('Matches Played Over The Years')
plt.xlabel('Year')
plt.ylabel('Number of Matches')
plt.show()
```

```python
performance_by_team=ipl_data.groupby(['batter','batsman_runs'])['total_runs'].sum().unstack().fillna(0)
performance_by_team=ipl_data.groupby(['bowler','player_dismissed','dismissal_kind'])['is_wicket'].sum().unstack().fillna(0)
sns.heatmap(performance_by_team,cmap='viridis')
plt.title('Player Performance by team')
plt.xlabel('Player')
plt.ylabel('Team')
plt.show()
```
